```python
from IPython.display import clear_output
import random

chips = 500

cards = {
    "2": [2, 4, 1],
    "3": [3, 4, 2],
    "4": [4, 4, 3],
    "5": [5, 4, 4], 
    "6": [6, 4, 5],
    "7": [7, 4, 6],
    "8": [8, 4, 7],
    "9": [9, 4, 8],
    "10": [10, 4, 9],
    "J": [10, 4, 10],
    "Q": [10, 4, 11],
    "K": [10, 4, 12],
    "A": [[1, 11], 4, 13]
}

class Deck(object):
    
    def draw(self, cards, firstDraw):
        selectedCards = []
        drawCount = None
        if firstDraw == True:
            drawCount = 2
      
        else:
            drawCount = 1
        while len(selectedCards) < drawCount:
            i = len(selectedCards)
            chosenCard = random.choice(cards.keys())
            while True:
                if self.cards[chosenCard][1] == 0:
                    chosenCard = random.choice(cards.keys())
                else:
                    break
            
            selectedCards.append(chosenCard)
            self.cards[selectedCards[i]][1] -= 1
            if firstDraw == False:
                selectedCards = selectedCards[0]

        return selectedCards
            
    
    def __init__(self, cards):
        self.cards = cards
        
        
        
class Hand(object): 
    
    def countHand(self, deck, auto): 
        handCount = 0
        aceSet = False
        for card in self.currentHand:
            if card == "A" and aceSet == False:
                
                if auto == True:
                    handCount += 1
                else:
                    while True:
                        try:
                            print "Your hand is %s" %(self.currentHand)
                            ace = int(raw_input("What would you like your ace to be counted as for this hand 1 or 11"))
                            aceSet = True
                        except:
                            print "That is not a number"
                            continue
                        else:
                            if ace == 1 or ace == 11:
                                handCount += ace
                                break
                            else:
                                print "This number is not 1 or 11"
                                continue
            else: 
                handCount += deck[card][0]
        
        return handCount 
    
        
    def __init__(self):
        self.currentHand = []

        
def makeABet(): 
    global chips
    while True:
        try:
            print "You currently have %s chips" %(chips)
            pot = int(raw_input("How much would you like to bet: "))
        except:
            print "Looks like you did not enter a number"
            continue
        else:
            if pot > chips:
                print "Sorry you do not have this much money to bet in your pot"
                continue
            else:
                print "You bet %s" %(pot)
                chips -= pot
                print "Remaining chips %s" %(chips)
                return pot
                break

        
def playerTurn():

    while True:
        try:
            val = raw_input("Hit or Stick: ").lower()
        except:
            print "Looks like you did not enter a legal command, please pick either Stick or Hit"
            continue
        else:
            if val == 'stick' or val == 'hit':
                return val
                break
            else:
                print "Looks like you did not enter a legal command!"
                continue

def restart(fullReset):
    global chips
    
    if fullReset == True:
        chips = 500
    else:
        while True:
            try:
                response = raw_input("Do you want to keep going? (Yes or No): ").lower()
            except:
                print "Looks like you did not enter a legal command, please pick either Yes or No"
                continue
            else:
                if response == 'yes' or response == 'no':
                    if response == "yes":
                        gameLoop()
                        break
                    else:
                        clear_output()
                        print "Time to cash in. You have %s chips" %(chips)
                        break
                else:
                    print "Looks like you did not enter a legal command! Only Yes or No"
                    continue
        
def compareHands(playersHand, dealersHand, deck):
    playersCount = 0
    dealersCount = 0
    for card in playersHand:
        playersCount += deck.cards[card][2]
    for card in dealersHand:
        dealersCount += deck.cards[card][2]
        
    if playersCount > dealersCount:
        return True
    else:
        return False
        
        
        
        
def gameLoop():
    global chips
    pot = 0
    playerScore = 0
    dealerScore = 0
    newHand = Hand()
    deck = Deck(cards)
    newHand.currentHand = deck.draw(deck.cards, True)
    print "Your first draw is %s giving you a score of %s" %(newHand.currentHand, newHand.countHand(deck.cards, False))
    if chips <= 0:
        print "Uh oh looks like you're out of chips, game restarting"
        restart(True)
    pot = makeABet()
    while True:
        clear_output()
        print "Your Current Hand is %s giving you a score of %s" %(newHand.currentHand, newHand.countHand(deck.cards, False))
        print "Your pot is %s and your remaining chips are %s" %(pot, chips)
        if playerTurn() == "hit":
            newHand.currentHand.append(deck.draw(deck.cards, False))
            handScore = newHand.countHand(deck.cards, False)
            if handScore > 21:
                print "Your hand is %s" %(newHand.currentHand)
                print "Bust! Game Over"
                playerScore = newHand.countHand(deck.cards, False)
                break
        else:
            playerScore = newHand.countHand(deck.cards, False)
            break
            
    dealersHand = Hand()
    dealersHand.currentHand = deck.draw(deck.cards, True)

    while True:
        
        print "Dealers Hand is %s giving them a score of %s" %(dealersHand.currentHand, dealersHand.countHand(deck.cards, True))
        handScore = dealersHand.countHand(deck.cards, True)
        if handScore <= 16 or (handScore < playerScore and playerScore <= 21):
            dealersHand.currentHand.append(deck.draw(deck.cards, False))
            handScore = dealersHand.countHand(deck.cards, True)
            print "Dealer Hits"
            if handScore > 21:
                print "Dealers Hand is %s giving them a score of %s" %(dealersHand.currentHand, dealersHand.countHand(deck.cards, True))
                print "Dealer is bust"
                dealerScore = handScore
                break
        else: 
            print "Dealer Sticks"
            dealerScore = handScore
            break
    print "Dealer scored %s" %(dealerScore)
    print "Player scored %s" %(playerScore)
    if playerScore > dealerScore and playerScore < 22 or dealerScore > 21:
        chips += (pot*2)
        print "Players Hand is %s" %(newHand.currentHand)
        print "Player Wins!"
        print "You have won %s and now have a %s chips" %(pot*2, chips)
    
    elif playerScore == dealerScore:
        if compareHands(newHand.currentHand, dealersHand.currentHand, deck) == True:
            chips += (pot*2)
            print "Players Hand is %s" %(newHand.currentHand)
            print "Player Wins!"
            print "You have won %s and now have a %s chips" %(pot*2, chips)
        else:
            print "Dealers Hand is %s" %(dealersHand.currentHand)
            print "Dealer Wins!"
            print "You have lost %s and now have %s chips" %(pot, chips)
        
    elif dealerScore > playerScore or playerScore > 21:
        print "Dealers Hand is %s" %(dealersHand.currentHand)
        print "Dealer Wins!"
        print "You have lost %s and now have %s chips" %(pot, chips)
        
    elif dealerScore > 21 and playerScore > 21:
        print "It's a tie, chips returned"
        chips += pot
    
    restart(False)
    
gameLoop()
    
```
